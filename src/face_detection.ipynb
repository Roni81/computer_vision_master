```python
img_resized = cv2.resize(img, (800, 600))
img_resized.shape
```

```python
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
```

```python
img_gray.shape
```

```python
cv2.imshow("people1.jpg",img_gray)

cv2.waitKey(0) # 키 입력 대기
cv2.destroyAllWindows() # 창 닫기 명령
```

```python
cv2.waitKey(1)
```

```python
face_detector = cv2.CascadeClassifier('/Users/sungminhong/Documents/Udemi/computer Vision Master/Cascades/haarcascade_frontalface_default.xml')
```

```python
detections = face_detector.detectMultiScale(img_gray)
```

```python
detections
```

```python
len(detections)
```

```python
for (x, y, w, h) in detections:
    # print(x, y, w, h)
    cv2.rectangle(img_resized, (x, y), (x+w, y+h), (0, 255, 0), 2)
                                                    # channel and thickness
cv2.imshow("Frame", img_resized)





cv2.waitKey(0) # 키 입력 대기
cv2.destroyAllWindows() # 창 닫기 명령
```
